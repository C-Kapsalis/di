```python
# Need to apply some temporary encoding to the target variable so as to create EDA graphs - did so by adding a temporary column in the
# training data, which we will drop before pre-processing so it is done properly later on (Label Encoding, training the corresponding class of sklearn
# on the training data and fitting it on the testing data).
def target_temp_encoding(target_col):
    if target_col == ' <=50K':
        return 0
    else:
        return 1
```

```python
train['target_temp'] = train['target'].apply(target_temp_encoding)
train
```

```python
# Creating a function that will encode 'hours-per-week' based on the common weekly workload regimes in the US
def working_week_type(hours_worked):
    if hours_worked < 30:
        return 'part_time'
    elif hours_worked <= 40:
        return 'national_standard'
    else:
        return 'long_week'
```

```python
# Creating a funtion that will take as input a dataframe (either 'train' or 'test' at this point), a column that we want to check to what 
# class each value of it corresponds ('target_col'), a categorization function in case it is a numeric column that we want to aggregate 
# so as to make translating results easier ('categorization_fnc'). In case the latter has been provided, we may also insert a name for it.
def freq_target_table(df, target_col, categorization_fnc=None, cat_col_name='agg_col', my_rotation=30):  # df will be either train or test + we will not use categorization_functions for some 'object' type categorical variables that need no further engineering
    
    # getting the absolute frequencies of each 'target' class for every single value of 'target_col'
    temp_1 = df.groupby(target_col)['fnlwgt'].count().reset_index().set_index(target_col)
    
    # getting the count of minority class samples ('>50K') for each single value
    temp_2 = df.groupby(target_col)['target_temp'].sum().reset_index().set_index(target_col)
    
    # putting the two previous together
    temp_fin = temp_1.join(temp_2).reset_index()
    temp_fin.columns = [target_col, 'freq', 'target_temp']

    # adding the new aggregation column should a categorization fnc exist
    if categorization_fnc != None:
        temp_fin[cat_col_name] = temp_fin[target_col].apply(categorization_fnc)
        
        # aggregating the absolute values based on the new categorization function
        agg_freqs = pd.pivot_table(data=temp_fin, index=cat_col_name, values='freq', aggfunc='sum').reset_index()
        agg_minority = pd.pivot_table(data=temp_fin, index=cat_col_name, values='target_temp', aggfunc='sum').reset_index()

        # finding relative values - first for the minority class sample participation
        rel_target_temp = []
        for wtype in agg_minority[cat_col_name]:  # labels are unique anyways, so there is no need to use the .unique() method
            total = agg_freqs[agg_freqs[cat_col_name] == wtype]['freq'].values[0]
            rel_target_temp.append(agg_minority[agg_minority[cat_col_name] == wtype]['target_temp'].values[0] / total)
        agg_minority['relative_target_temp'] = rel_target_temp

        # now the relative frequencies of each label
        rel_freqs = []
        for wtype in agg_minority[cat_col_name]:  # labels are unique anyways, so there is no need to use the .unique() method
            total = agg_freqs[agg_freqs[cat_col_name] == wtype]['freq'].values[0]
            rel_freqs.append((total - agg_minority[agg_minority[cat_col_name] == wtype]['target_temp'].values[0]) / total)
        agg_minority['rel_freq'] = rel_freqs

        # constructing the final plot
        fig, ax = plt.subplots(figsize=(13, 8))
        ax.bar(x=agg_minority[cat_col_name].unique(), height=agg_minority['relative_target_temp'], bottom=0, label='> $50K', color='black')
        ax.bar(x=agg_minority[cat_col_name].unique(), height=agg_minority['rel_freq'], bottom=agg_minority['relative_target_temp'], label='<= $50K', color='red')
        ax.set_xticklabels(ax.get_xticklabels(), rotation=my_rotation)
        ax.legend()
        plt.show()

    else: 
        cat_col_name = target_col
        
        # aggregating the absolute values based on the new categorization function
        agg_freqs = pd.pivot_table(data=temp_fin, index=cat_col_name, values='freq', aggfunc='sum').reset_index()
        agg_minority = pd.pivot_table(data=temp_fin, index=cat_col_name, values='target_temp', aggfunc='sum').reset_index()

        # finding relative values - first for the minority class sample participation
        rel_target_temp = []
        for wtype in agg_minority[cat_col_name]:  # labels are unique anyways, so there is no need to use the .unique() method
            total = agg_freqs[agg_freqs[cat_col_name] == wtype]['freq'].values[0]
            rel_target_temp.append(agg_minority[agg_minority[cat_col_name] == wtype]['target_temp'].values[0] / total)
        agg_minority['relative_target_temp'] = rel_target_temp

        # now the relative frequencies of each label
        rel_freqs = []
        for wtype in agg_minority[cat_col_name]:  # labels are unique anyways, so there is no need to use the .unique() method
            total = agg_freqs[agg_freqs[cat_col_name] == wtype]['freq'].values[0]
            rel_freqs.append((total - agg_minority[agg_minority[cat_col_name] == wtype]['target_temp'].values[0]) / total)
        agg_minority['rel_freq'] = rel_freqs

        # constructing the final plot
        fig, ax = plt.subplots(figsize=(13, 8))
        ax.bar(x=agg_minority[cat_col_name].unique(), height=agg_minority['relative_target_temp'], bottom=0, label='> $50K', color='black')
        ax.bar(x=agg_minority[cat_col_name].unique(), height=agg_minority['rel_freq'], bottom=agg_minority['relative_target_temp'], label='<= $50K', color='red')
        ax.set_xticklabels(ax.get_xticklabels(), rotation=my_rotation)
        ax.legend()
        plt.show()
        

freq_target_table(train, 'hours-per-week', working_week_type, 'week-type')
```

### Feature Engineering


#### Feature Selection: 
We have already dropped 'fnlwgt' from the training and test sets because we are trying to predict whether someone makes more than $50K/year in the US by analyzing their socio-economic attributes, irrespective of the state they are based in. The dataset's documentation, however, states that 'fnlwgt' estimates are constructed on the state level. Different states may bear significantly different socio-economic characteristics. As such, we would keep it only if we also had some state-level information, at least what the state of each individual mentioned is.

Furthermore, the information given by 'education' and 'education-num' seems redundant, and the way the two variables influence the training of classification models is different. From common experience, we know that what matters most is the educational certifications that a person holds. As for people with less than a high-school diploma, just knowing the highest level of school education they received is enough. 

So we dropped  'education-num' variable and kept the categorical 'education' variable that we would later on perform one-hot-encoding upon. This will help prevent introducing any bias to the model, as we saw in the training set before that people with higher academic qualifications were high earners at a higher rate than the re, and we want an objective estimation of how important this factor actually is.



#### Aggregating information:
We performed the aforementioned operation on the ‘education’ column by extracting the cardinal number of the last class attended by relevant individuals, whom we then grouped together based on the following segmentation: 
- 	Elementary school – attended up to the 5th grade of school - •	Junior high school - up to the 8th grad- 
•	High school - up to the 12th gr

```python
pca = PCA()  # we will retain all initial columns
PCs = ['PC'+str(i+1) for i in range(X_train.shape[1])]

X_train_scores = pd.DataFrame(pca.fit_transform(X_train.values), columns=PCs)
X_test_scores = pd.DataFrame(pca.transform(X_test.values), columns=PCs)
```

```python
tost2 = pd.DataFrame(data=pca.components_, columns = X_train.columns, index = PCs) 
tost2.head()
```

```python
# retrieving the three initial features that influenced the values of a PC the most, i.e. they had the highest pc scores
three_most_important = pd.DataFrame({'first': [None]*len(PCs), 'second': [None]*len(PCs), 'third': [None]*len(PCs)}, index=PCs)
pc_to_features_dict = dict()
for pc in PCs:
    pc_to_features_dict[pc] = list(tost2.loc[pc, :].abs().transpose().sort_values(axis=0, ascending=False).index[:3])
    
```

```python
# creating a function so that if I give it a list of principal components names (in the form of PC{int}), it will revert with the frequency 
# distribution of the features of the original dataset that do the best job explaining the values of each PC. 
# This way, when implementing explainable ai techniques on the fully pre-processed dataset, upon finding the most important features of that 
# (which resemble principal components), I can trace back to the original dataset's features contributing to them. 

def pcs_to_features(pc_names_list):
    working_str = ''
    for pc in pc_names_list:
        working_str += '|'.join(pc_to_features_dict[pc]) + '|'
    unique_feature_names = list(set(working_str.split('|')[:-1]))  # to prevent also including an empty string originating from the concluding '|' character in working_str
    freqs = []
    for i in range(len(unique_feature_names)):
        freqs.append(working_str.count(unique_feature_names[i]))
        
    # mapping each underlying feature to its absolute frequency in a dedicated df that can used for plotting and reporting 
    feature_df = pd.DataFrame(data={'feature': unique_feature_names, 'frequency': freqs})
    feature_df.sort_values(by='frequency', ascending=False, inplace=True)  # I need to sort it inplace so as to get that order in matplotlib's bar chart as well
    
    return feature_df

def plot_feature_df(feature_df):
    # plotting the underlying features in a bar chart whose bars are ordered in descending order based on each feature's frequency
    fig, ax = plt.subplots(figsize=(13, 8))
    ax.bar(x=feature_df['feature'], height=feature_df['frequency'])
    ax.set_xticklabels(ax.get_xticklabels(), rotation=30)
    plt.show()

```

```python
def pcs_to_features_by_effect_sign(pc_names_list):
    # given a list of PCs, some features that positively impact a PC might negatively impact others. If I apply it to just one PC (e.g. ['PC1']), 
    # I will find the main way of contribution to its values (positive/negative) of its most contributing features, so by combining that to the 
    # local SHAP scores I retrieve for that given PC, I can get a sense of how the change in the original features values affect predictions.
    
    
    # same as before but for 5 features at a time
    pc_to_features_dict5 = dict()
    for pc in PCs:
        pc_to_features_dict5[pc] = list(tost2.loc[pc, :].abs().transpose().sort_values(axis=0, ascending=False).index[:5])

    t_list = []
    for pc in pc_names_list:
        t_list += pc_to_features_dict5[pc]
    t_list = list(set(t_list))  # to eliminate duplicates
    running = tost2.loc[pc_names_list][t_list]
    
    sign_count = dict()
    for col in running.columns: 
        sign_count[col] = [len(running[col][running[col] < 0].index), len(running[col][running[col] >= 0].index)]
    negative_occurrences = [el[0] for el in sign_count.values()]
    positive_occurrences = [el[1] for el in sign_count.values()]
    
    fig, ax = plt.subplots(figsize=(13, 8))
    ax.bar(x=list(sign_count.keys()), height=negative_occurrences, bottom=0, label='negative occurrencies', color='black')
    ax.bar(x=list(sign_count.keys()), height=positive_occurrences, bottom=negative_occurrences, label='positive occurrencies', color='red')
    ax.set_xticklabels(ax.get_xticklabels(), rotation=30)
    ax.legend()
    plt.show()
```

```python
def clustered_mean_pos_and_neg_score(pc_names_list):
    # inspired from matplotlib's documentation: https://matplotlib.org/stable/gallery/lines_bars_and_markers/barchart.html 
    pc_to_features_dict5 = dict()
    for pc in PCs:
        pc_to_features_dict5[pc] = list(tost2.loc[pc, :].abs().transpose().sort_values(axis=0, ascending=False).index[:5])

    t_list = []
    for pc in pc_names_list:
        t_list += pc_to_features_dict5[pc]
    t_list = list(set(t_list))  # to eliminate duplicates
    running = tost2.loc[pc_names_list][t_list]
    
    avg_per_sign = dict()
    for col in running.columns: 
        avg_per_sign[col] = [running[col][running[col] < 0].mean(), running[col][running[col] >= 0].mean()]
    negative_occurrences = [el[0] for el in avg_per_sign.values()]
    positive_occurrences = [el[1] for el in avg_per_sign.values()]
    features = list(avg_per_sign.keys())
    tot_dict = {'neg_occs': negative_occurrences, 'pos_occs': positive_occurrences}

    x = np.arange(len(running.columns))  # the label locations
    width = 0.25  # the width of the bars
    multiplier = 0

    fig, ax = plt.subplots(figsize=(14,8))

    for attribute, measurement in tot_dict.items():
        offset = width * multiplier
        rects = ax.bar(x + offset, measurement, width, label=attribute)
        ax.bar_label(rects, padding=3)
        multiplier += 1

    # Add some text for labels, title and custom x-axis tick labels, etc.
    ax.set_ylabel('Score')
    ax.set_title('Average Positive and Negative Score per feature for the top 5 PCs')
    ax.set_xticks(x + width, features)
    ax.legend(loc='upper left', ncols=2)
    ax.set_ylim(-0.5, 0.5)
    ax.set_xticklabels(ax.get_xticklabels(), rotation=90)
    ax.axhline(0, color='black')

    plt.show()
    
```

```python
param_grid = {'penalty': [None, 'l1', 'l2'],  # regularization similar to ordinary linear regression; l1 adds an l1 (i.e. absolute weight) penalty term, whereas l2 adds an l2 (i.e. squared weight) so as to constrain the magnitude of weights/estimated model coefficients - the higher their magnitude, the higher their variance, so we try to limit that by voluntarily introducing some bias to the model with the goal of reducing its probability of overfitting 
              'solver': ['sag', 'saga', 'liblinear'],
              # from sklearn's documentation we see that sag & saga are faster for larger datasets, and 'newton-cholesky' is good when n_samples>>n_features and we have one-hot-encoded categorical features with rare categories, like in our case
              'C' : np.logspace(-4, 4, 20),  # lower values - higher regularization strength, i.e. higher penalty for misclassifications
              'class_weight' : [None, 'balanced']  
              }

log_reg_grid_cv = GridSearchCV(estimator=LogisticRegression(random_state=0, max_iter = 500),  # try a higher max iter value
                         #            n_iter = 35,
                         #  param_distributions = param_grid,
                                     param_grid = param_grid,
                           n_jobs=-1,
                           cv=5  # inserting an integer in the classification case ensures that stratified k-fold cross validation takes place 
                           )

log_reg_grid_cv.fit(X_train_smpl.values, y_train_smpl.values.ravel())


# even without checking all parameter options due to difficulty converging, we got much better results with log reg
```
